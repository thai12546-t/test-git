## 1. Nạp dự liệu iris vào bộ nhớ và hiển thị thông tin về số thuộc tính (dimensionality), số objects (size).

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd 

break_cancer = load_breast_cancer()
df = pd.DataFrame(data=break_cancer.data, columns= break_cancer.feature_names)
print(df.shape)
```

## 2. Đặt nhãn (label) cho dữ liệu đã nạp và hiển thị số lớp (classes)

```python
labels = break_cancer.target_names
feature_names = break_cancer.feature_names

print(f'Number of classes : {len(labels)}')
print(labels)

print(f'Number of Features : {len(feature_names)}')
print(feature_names)
```

## 3. Tách dữ liệu thành hai phần: training data(70%) và test data (30%)

```python
from sklearn.model_selection import train_test_split 

X = df[feature_names]
Y = break_cancer.target
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, train_size=0.7, test_size=0.3)
print(X_train.shape, Y_train.shape)
print(X_test.shape, Y_test.shape)

print(X_train.head(3))
print(X_test.head(3))
```

## 4. Chuẩn hóa dữ liệu với StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

## 5. Tạo và huấn luyện mô hình với training data

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=2, random_state=0)
clf.fit(X_train, Y_train)
```

## 6. Test và đánh giá độ chính xác của mô mình, yêu cầu hiển thị ma trận confusion matrix, accuracy, f-score

```python
Y_pred = clf.predict(X_test)
Y_True = Y_test
print(f'Input : {str(X_test[:9])}')
print(f'Output : {str(Y_pred[:9])}')
print(f'True values : {str(Y_True[:9])}')
```

```python
from sklearn.metrics import confusion_matrix 

print(f'Confusion Matrix : {confusion_matrix(Y_True, Y_pred)}')
```

```python
from sklearn.metrics import accuracy_score, f1_score

print(f'Accuracy : {accuracy_score(Y_True, Y_pred)}')
print(f'F1 Score : {f1_score(Y_True, Y_pred, average=None)}')
```

## 7. Trực quan hóa cây quyết định

```python
import graphviz
from sklearn import tree

dot_data = tree.export_graphviz(clf, feature_names=break_cancer.feature_names, class_names=labels, filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph
```
